# Creating the Dictionary thing

```python
last_name_df = data.copy()
```

```python
def last_name(s):
    s_list = s.split(' ')
    return s_list[len(s_list)-1]
last_name_df['name'] = last_name_df['name'].apply(last_name)
```

```python
last_name_df
```

```python
data
```

```python
last_name_dict = {}
for row in last_name_df.values:
    # row gives list of the values of the row i.e. ['Bellamy' ' Sc.D.' 'Associate Professor of Biostatistics' 'bellamys@mail.med.upenn.edu']
    row = row.tolist() # instead of numpy array
    if row[0] in last_name_dict.keys():
        last_name_dict[row[0]] = [last_name_dict[row[0]],row[1:]]
    else:
        last_name_dict[row[0]] = row[1:]
```
